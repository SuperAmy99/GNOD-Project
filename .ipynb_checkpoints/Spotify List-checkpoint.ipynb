## Web Scraping: Create a list of top_200_songs 

### Web Scraping Popvortex

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# find url and store it in a variable
url = "https://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
# download html with a get request
response = requests.get(url)
response.status_code
```

```python
# parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# check that the html code looks like it should
# soup
```

```python
title = []
for t in soup.select('div.chart-content.col-xs-12.col-sm-8 > p > cite'):
    title.append(t.get_text())
# print(title)
```

```python
artist = []
for a in soup.select('div.chart-content.col-xs-12.col-sm-8 > p > em'):
    artist.append(a.get_text())
# print(artist)
```

```python
popvortex_list = pd.DataFrame({"title":title,
                       "artist":artist,
                      })
popvortex_list
```

```python
from random import randint
from time import sleep

def get_playlist_tracks(playlist_id):
    results = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = results['items']
    while results['next']!=None:
        results = sp.next(results)
        tracks = tracks + results['items']
        sleep(randint(1,3000)/1000) # respectful nap
    return tracks
```

```python
# define a function to get more playlist
def get_playlist_dataframe(sp, playlist_id):
    def expand_list_dict2(row):
        df = json_normalize(row['track.artists'])
        df['song_id'] = row['track.id']
        return df

    def fetch_audio_features(sp, id_list):
        chunks = [(i, i+100) for i in range(0, len(id_list), 100)]
        audio_features_list = []

        for chunk in chunks:
            id_list100 = id_list[chunk[0]:chunk[1]]

            # Filter out None or invalid track IDs
            id_list100 = [track_id for track_id in id_list100 if track_id is not None]

            if id_list100:
                audio_features_list += sp.audio_features(id_list100)
                sleep(randint(1, 3000) / 1000)
            else:
                # Append None for each track ID that is None
                audio_features_list += [None] * len(id_list100)

        return audio_features_list

    # Fetch all tracks from the playlist
    results = sp.user_playlist_tracks("spotify", playlist_id)
    tracks = results['items']

    while results['next'] is not None:
        results = sp.next(results)
        tracks += results['items']
        sleep(randint(1, 3000) / 1000)  # respectful nap

    # Create a DataFrame for the tracks
    tracks_df = json_normalize(tracks)

    # Create a DataFrame for the artists
    tracks_df['artists_dfs'] = tracks_df.apply(expand_list_dict2, axis=1)
    artist_df = pd.DataFrame(columns=['href', 'id', 'name', 'type', 'uri'])

    for mini_df in tracks_df['artists_dfs']:
        artist_df = pd.concat([artist_df, mini_df], axis=0)

    # Merge the two dataframes, keeping only the fields needed
    df_merged = pd.merge(left=tracks_df,
                        right=artist_df,
                        how='inner',
                        left_on='track.id',
                        right_on='song_id')

    # Select the final columns
    df_final = df_merged[['track.name', 'name', 'song_id']]

    # Fetch audio features for the tracks
    id_list = df_final['song_id'].tolist()
    audio_features_list = fetch_audio_features(sp, id_list)

    # Create a DataFrame for the audio features
    audio_features_df = pd.DataFrame(audio_features_list)

    # Drop duplicates in the audio features dataframe
    audio_features_df.drop_duplicates(inplace=True)

    # Merge the final dataframe with the audio features dataframe
    df_w_audio_ft = pd.merge(left=df_final,
                             right=audio_features_df,
                             how='inner',
                             left_on='song_id',
                             right_on='id')

    return df_w_audio_ft
```

```python
# Usage
playlist_id_2 = "6FKDzNYZ8IW1pvYVF4zUN2"
df_playlist_2 = get_playlist_dataframe(sp, playlist_id_2)
```

## Unsupervised Learning: K-means clustering

```python
import numpy as np
numerical = df.select_dtypes(include=np.number)
numerical.shape
```

```python
from sklearn.preprocessing import StandardScaler
X_prep = StandardScaler().fit_transform(numerical)


# Now, all features will have the same weight.
X = pd.DataFrame(X_prep,columns=numerical.columns)
X
```

```python
X.to_csv('standard_cluster.csv', index=False)
```

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

K = range(2, 20)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
from sklearn.metrics import silhouette_score
K = range(2, 20)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X)
    silhouette.append(silhouette_score(X, kmeans.predict(X)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Slhouette score showing the optimal k')
```

```python
# according to the graph above, I choose 8 clusters
```

```python
# from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=8, random_state=1234)
kmeans.fit(X)
```

```python
# Predicting / assigning the clusters:
clusters = kmeans.predict(X)

# Check the size of the clusters
pd.Series(clusters).value_counts().sort_index()
```

## Build song recommendation function


```python
# Get the user's input song
user_input = input("Please Enter a song: ")

# Check if the user's input song is in the Hot 200 (case-sensitive)
if (hot_list['title'].str.lower() == user_input.lower()).any():
    # If it's in the Hot 200, recommend another Hot 200 song
    recommended_hot_song = hot_list[hot_list['title'] != user_input].sample(1)
    print("You seem to enjoy hot songs! Here is a recommendation for you:")
    print(f"{recommended_hot_song['title'].values[0]} by {recommended_hot_song['artist'].values[0]}")

else:
    # If it's not in the Hot 200, fetch audio features from the Spotify API
    # and use the clustering model to recommend a song from the same cluster

    # Assume you have a pickled KMeans model saved, replace 'kmeans_model.pkl' with the actual file path
    with open('kmeans.pkl', 'rb') as model_file:
        kmeans_model = pickle.load(model_file)
    with open('scaler.pkl', 'rb') as scaler_file:
        scaler = pickle.load(scaler_file)

    # Find the song URI using Spotify search
    results = sp.search(q= user_input, type="track")

    if results['tracks']['items']:
        song_uri = results['tracks']['items'][0]['uri'].split(':')[-1]

        # Find song features using Spotify audio_features
        song_features = sp.audio_features(song_uri)

        if song_features:
            # Create a DataFrame for the audio features
            user_song_features = pd.DataFrame(song_features)
            
            # Select only numerical columns
            numerical_columns = user_song_features.select_dtypes(include=['number']).columns
            user_song_features_numeric = user_song_features[numerical_columns]

            user_song_scaled = scaler.transform(user_song_features_numeric)

            # Predict the cluster for the user's song
            user_cluster = kmeans_model.predict(user_song_scaled)

#             print("Predicted Cluster for the User's Song:", user_cluster[0])

              # recommend same cluster song
            recommended_cluster_song = spotify_with_cluster[spotify_with_cluster['cluster'] == user_cluster[0]].sample(1)

            if not recommended_cluster_song.empty:
                print("You might also like:")
                print(recommended_cluster_song['song_and_artist'].to_string(index=False))

            else:
                print("No recommendations available for the user's song cluster.")
        else:
            print("Unable to fetch audio features for the user's song.")
    else:
        print("Song not found in Spotify. Please enter another song.")
```
