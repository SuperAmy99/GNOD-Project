```python
# just for recommendator


import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import random
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pickle
```

```python
secrets_file = open("secrets.txt","r")
string = secrets_file.read()
```

```python
secrets_dict={}
for line in string.split('\n'):
    if len(line) > 0:
        #print(line.split(':'))
        secrets_dict[line.split(':')[0]]=line.split(':')[1].strip()
```

```python
#Initialize SpotiPy with user credentials
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=secrets_dict['clientid'],
                                                           client_secret=secrets_dict['clientsecret']))
```

```python
# first import the current hot list

hot_list = pd.read_csv('hot_list.csv')
hot_list
```

```python
playlist_clustered = pd.read_csv('spotify_with_cluster.csv')
playlist_clustered
```

```python
# Get the user's input song
user_input = input("Please Enter a song: ")

# Check if the user's input song is in the Hot 200 (case-sensitive)
if hot_list['title'].str.lower().str.contains(user_input.lower()).any():
    # If it's in the Hot 200, recommend another Hot 200 song
    recommended_hot_song = hot_list[hot_list['title'] != user_input].sample(1)
    print("You seem to enjoy hot songs! Here is a recommendation for you:")
    print(f"{recommended_hot_song['title'].values[0]} by {recommended_hot_song['artist'].values[0]}")

else:
    # If it's not in the Hot 200, fetch audio features from the Spotify API
    # and use the clustering model to recommend a song from the same cluster

    # Assume you have a pickled KMeans model saved, replace 'kmeans_model.pkl' with the actual file path
    with open('kmeans.pkl', 'rb') as model_file:
        kmeans_model = pickle.load(model_file)
    with open('scaler.pkl', 'rb') as scaler_file:
        scaler = pickle.load(scaler_file)

    # Find the song URI using Spotify search
    results = sp.search(q="track:" + user_input, type="track")
    song_uri = results['tracks']['items'][0]['uri'].split(':')[-1]

    # Find song features using Spotify audio_features
    song_features = sp.audio_features(song_uri)

    # Assuming you have a DataFrame named 'spotify_data' with audio features and a column 'cluster'
    # Replace 'title' with the actual column name
    user_song_features = pd.DataFrame(song_features)[['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'valence']]
    user_song_scaled = scaler.transform(user_song_features)

    # Predict the cluster for the user's song
    user_cluster = kmeans_model.predict(user_song_scaled)

    # Assuming you have a DataFrame named 'playlist_clustered' with columns 'title', 'artist', 'url', and 'cluster'
    # Replace 'title' with the actual column name
    recommended_cluster_song = playlist_clustered[playlist_clustered['cluster'] == user_cluster[0]].sample(1)
    print("Here's what I found for you!")
    print(recommended_cluster_song[['title', 'artist', 'url']].to_string(index=False))
```
